```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from xbbg import blp
```

```python
def GetRet(Liste, Start = "20150101"):
    Start = "20150101"
    End = pd.Timestamp.today().strftime("%Y%m%d")

    Res = blp.bdh(tickers= Liste,start_date = Start , end_date = End)
    Res.columns = [x[0] for x in Res.columns]
    Ret = (np.log(Res) - np.log(Res.shift(1))).iloc[1:,:]
    #Ret = Ret.T.mean()
    return Ret

def Plot(Ret):
    Ret = Ret + 1
    if type(Ret) == type(pd.Series(0)):
        Ret[0] = 100
    else:
        Ret.iloc[0,0] = 100
    plt.plot(Ret.cumprod())
```

```python
CurrList = ["EURUSD Curncy", "JPYUSD Curncy", "GBPUSD Curncy", "AUDUSD Curncy", "CHFUSD Curncy", "CADUSD Curncy", "MXNUSD Curncy", "BRLUSD Curncy", "NOKUSD Curncy", "SEKUSD Curncy", "RUBUSD Curncy", "SGDUSD Curncy", "KRWUSD Curncy", "TWDUSD Curncy"]

```

```python
df = GetRet(CurrList, Start = "20150101")
```
