```python
Pointer = {}

for x in range(252, len( Vol )):
    if x%20 == 0:
        Pointer[Vol.index[x]] = np.array(rates[[CurrData[x]["Central bank rate"] for x in CurrData.keys()]].iloc[x]) / np.array(Vol[[x for x in CurrData.keys()]].iloc[x])
    else:
        Pointer[Vol.index[x]] = np.nan
        
```
