# Safe Haven Basket

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from xbbg import blp
```

```python
def GetRet(Liste, Start = "20150101", End = pd.to_datetime("today").strftime("%Y%m%d"), Returns = True):
    Res = blp.bdh(tickers= Liste,start_date = Start , end_date = End)
    Res.columns = [x[0] for x in Res.columns]
    Ret = (np.log(Res) - np.log(Res.shift(1))).iloc[1:,:]
    if Returns == True:
        Ret = Ret.T.mean()
    return Ret

def GetBench(Index, Start = "20150101", End = pd.to_datetime("today").strftime("%Y%m%d")):
    Res = blp.bdh(tickers= Index + " Index",start_date = Start , end_date = End)
    Ret = (np.log(Res) - np.log(Res.shift(1))).iloc[1:,:]
    Ret.columns = [Index]
    return Ret

def Plot(Ret):
    print("Sharpe", np.sqrt(252) * Ret.mean()/Ret.std())
    Ret = Ret + 1
    if type(Ret) == type(pd.Series(0)):
        Ret[0] = 100
    else:
        Ret.iloc[0,0] = 100
    plt.plot(Ret.cumprod())
```

```python
SHListe = ["JPYUSD Curncy", "GCA Comdty", "SIA Comdty", "XTC5 GY Equity"]
EQDSHListe = ["S5UTIL Index", "S5CONS Index", "S5HLTH Index"]
```

```python
Plot(GetRet(SHListe, Start = "20200101"))
```

```python
Plot(GetRet(EQDSHListe, Start = "20200101"))
```

```python
Plot(GetRet( ["S5UTIL Index"], Start = "20200101"))
Plot(GetRet( ["S5CONS Index"], Start = "20200101"))
Plot(GetRet( ["S5HLTH Index"], Start = "20200101"))
plt.legend(["Utilities", "Consumers Staples", "Health Care"])
plt.show()
```

```python
Sector = ["S5ENRS Index", "S5FINL Index", "S5MATR Index", "S5INDU Index", "S5HLTH Index", "S5TELS Index", "S5CONS Index", "S5COND Index", "S5INFT Index", "S5UTIL Index", "S5RLST Index"]
```

```python
Test = GetRet(Sector, Start="20200101", Returns = False)
```

```python
Test.iloc[0,:] = 1
```

```python
Test = Test + 1
```

```python
Test.iloc[0,:] = 100
```

```python
Test = Test.cumprod()
```

```python
plt.plot(Test)
plt.legend([x.split(" ")[0] for x in Test.columns])
plt.rcParams["figure.figsize"] = (500,10)
plt.show()
```
