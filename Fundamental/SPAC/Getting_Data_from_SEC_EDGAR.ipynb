# Getting Data

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
dfAll = None

for x in range(1,4001, 100):
    print(x)
    url = "http://www.sec.gov/cgi-bin/srch-edgar?text=ASSIGNED-SIC%20%3D%206770&start=" + str(x) + "&count=100&first=1994&last=2020"
    df = pd.read_html(url)[4]
    #if len(df) < 100:
    df = df[["html" in x for x in df.Format]]
    df.index = [x for x in range(len(df))]
    
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    table = soup.findAll('table')[4]
    links = []
    for tr in table.findAll("tr"):
        trs = tr.findAll("td")
        for each in trs:
            try:
                link = each.find('a')['href']
                if link[-4:] == ".htm":
                    links.append(link)
            except:
                pass

    df['Link'] = links
    
    CIKList = []
    for n in range(len(df)):
        url2 = "https://www.sec.gov" + df.Link[n]
        response2 = requests.get(url2)
        soup2 = BeautifulSoup(response2.text, 'html.parser')
        CIKList.append(soup2.text.split("CIK: ")[1][:10])
    
    df["CIK"] = CIKList
    
    if dfAll is None:
        dfAll = df
    else:
        dfAll = pd.concat((dfAll, df), axis = 0)
```
