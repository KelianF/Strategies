```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url = "http://www.sec.gov/cgi-bin/srch-edgar?text=ASSIGNED-SIC%20%3D%206770&start=1&count=100&first=1994&last=2020"
df = pd.read_html(url)[4]
```

```python
df
```

```python
response = requests.get(url)
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
table = soup.findAll('table')[4]
```

```python
table
```

```python
links = []
for tr in table.findAll("tr"):
    trs = tr.findAll("td")
    for each in trs:
        try:
            link = each.find('a')['href']
            if link[-4:] == ".htm":
                links.append(link)
        except:
            pass

df['Link'] = links
```
